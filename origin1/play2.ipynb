```python
# this is bs ...

from random import random, seed
seed = 500

def ReLU(v):
    return max(0, v)

class Neuron:
    def __init__(self, connections:list=None) -> None:
        self.connections = connections if connections else []
        self.v = 0
        
    def add_charge(self, v):
        self.v += ReLU(v)
        
        for c in self.connections:
            c.target.add_charge(self.v * c.w)
        
        


class Synapse:
    def __init__(self, target:Neuron, w=None) -> None:
        self.target = target
        self.w = w if w != None else 2*(0.5-random())
    
 
O, L1, L2, I1, I2 = None, None, None, None, None
    
def instantiate(i1_w, i2_w, l1_weights, l2_weights):
    global O, L1, L2, I1, I2
    
    O = Neuron()
    L2 = [Neuron([Synapse(O, w)]) for w in l2_weights]
    L1 = [Neuron([Synapse(n, w) for n in L2]) for w in l1_weights]

    I2 = Neuron([Synapse(n, i2_w) for n in L1])
    I1 = Neuron([Synapse(n, i1_w) for n in L1])
    
def simulate(charge1, charge2):
    I1.add_charge(charge1)
    I2.add_charge(charge2)
    return O.v
    
    


i1w = 1
i2w = -3
l1w = [w/10 for w in [1, 2, -5, -1, -8, 2]]
l2w = [w/10 for w in [4, 3, -3, -6, -4, 2, -6]]
instantiate(i1w, i2w, l1w, l2w)

n = 100
X = [i for i in range(-n, n, 1)]
Y = [i for i in range(-n, n, 1)]
results = []
for x in X:
    for y in Y:
        r = simulate(x, y)
        results.append({"x": x, "y": y, "z": r})

import matplotlib.pyplot as plt
%matplotlib ipympl

# ax = plt.axes(projection="3d")
# for r in results:
#     ax.scatter(r['x'], r['y'], r['r'])


```

```python
import plotly.io as pio
pio.renderers.default = "notebook_connected"
import plotly.express as px
import pandas as pd

def plot_vectors(vectors):
    color_values = [random() for v in vectors]
    opacity = 0.9
    title = "Title"; text = "Text"
    color_scale = [(0, "green")] # great
        
    mid_value=0.6; bad_value=None
    mid_normalized = mid_value / max(color_values)
    color_scale.append( (min(mid_normalized, 1), "yellow") ) # in range
    if bad_value != None:
        bad_normalized = bad_value/max(color_values)
        color_scale.append( ( min(1, bad_normalized), "#7D3C98") ) # gets bad
    if mid_normalized < 1 or (bad_value != None and bad_normalized < 1):
        color_scale.append((1, "red"))

    df = pd.DataFrame({"colors": color_values,
                        "x": [v['x'] for v in vectors],
                        "y": [v['y'] for v in vectors],
                        "z": [v['z'] for v in vectors]})
    fig = px.scatter_3d(df,
            x="x", y="y", z="z",
            color='colors',
            opacity=opacity,
            color_continuous_scale=color_scale,
            title=title+text,
        )
    fig.show()
```

```python

def foo(x:int, y:int):
    try: 
        return max(0, x)-max(x, y)
    except: return 0

def iterate(a, b, s):
    i = a
    while i < b:
        yield(float(i))
        i += s

results = []
n, step = 10, 1/3
for x in iterate(-n, n, step):
    for y in iterate(-n, n, step):
        results.append({"x": x, "y": y, "z": foo(x, y)})

plot_vectors(results)
```
